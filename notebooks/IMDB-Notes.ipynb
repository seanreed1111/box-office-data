```python
title_nums = ["tt0324216"] #The Texas Chainsaw Massacre
```

```python
def make_urls():
    base_url = "http://www.imdb.com/title/"
    urls = []
    
    for title in title_nums:
        urls.append(base_url + title + '/')
    return urls
```

```python
def my_count():
    n = 1000
    while True:
        yield n
        n += 1
    
numbers = my_count()
```

```python
def start(my_session = None):
    
    urls = make_urls()
    print('Urls', urls)
    
    for url in urls:
        try:
            r = my_session.get(url)
            print("request headers",r.request.headers)
            print("response headers",r.headers)
        except Exception as e:
            print("accessing url", url)    
            print('Exception encountered at position 1:', e)
            
    return r, urls
```

```python
def write(response, **kwargs):

    print('writing file from..{} '.format({response.url}))  
    filename =  response.url + ".html"
    with open(filename, 'wb') as f:
        f.write(response.content)
        f.write(response.url.encode('utf-8'))
        print('saved file %s' % filename)
```

```python
with Session() as s:
    try:
        start_tuple = start(s)
    except Exception as err:
        print('Exception encountered at position 2:', err)

```
